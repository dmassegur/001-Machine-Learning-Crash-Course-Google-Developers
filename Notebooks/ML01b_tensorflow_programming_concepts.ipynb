TensorFlow Programming Concepts:

```
from __future__ import print_function

%tensorflow_version 1.x
import tensorflow as tf
import matplotlib.pyplot as plt # Dataset visualization.
import numpy as np              # Low-level numerical Python library.
import pandas as pd             # Higher-level numerical Python library.

# Create a graph:  ## a graph is an operaiton (like a function)
g = tf.Graph()

# Establish the graph as the "default" graph:
with g.as_default():
  # Assemble a graph consisting of the following three operations:
  #   * Two tf.constant operations to create the operands.
  #   * One tf.add operation to add the two operands.
  x = tf.constant(8, name='x_const')
  y = tf.constant(5, name='y_const')
  my_sum = tf.add(x, y, name='x_y_sum')

  # Now create a session.
  # The session will run the default graph.
  with tf.Session() as sess:
    print(my_sum.eval())  # no need to initialise because there are no variables
    print(my_sum)
```

Working with variables

```
# Create a graph:  ## a graph is an operaiton (like a function)
g = tf.Graph()

# Establish the graph as the "default" graph:
with g.as_default():
  # Assemble a graph consisting of the following three operations:
  #   * Two tf.constant operations to create the operands.
  #   * One tf.add operation to add the two operands.
  x = tf.constant(8, name='x_const')
  y = tf.constant(5, name='y_const')
  my_sum = tf.add(x, y, name='x_y_sum')

  # Defining a variable
  z = tf.Variable([0])  # to initialise a variable
  z = z.assign([4]) # to modify a variable

  # Adding it to the sum:
  my_sum = tf.add(my_sum, z, name='x_y_z_sum')

  # Now create a session.
  # The session will run the default graph.
  with tf.Session() as sess:
    initialization = tf.global_variables_initializer()  # we need to initialise the variables
    print(z.eval())
    print(my_sum.eval()) 
    print(my_sum)
```
